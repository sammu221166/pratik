```python
import pandas as pd
from functools import reduce

# 1. Create the DataFrame
data = {
    'total_bill': [23.45, 35.70, 10.50, 42.30, 18.20, 27.35],
    'tip': [3.50, 5.20, 2.00, 6.50, 1.80, 5.75],
    'day': ['Sun', 'Sun', 'Mon', 'Mon', 'Tue', 'Tue']
}

df = pd.DataFrame(data)

# 2. Sum of total bill for each day
bill_sum_per_day = df.groupby('day')['total_bill'].sum()
print("Sum of total bill per day:\n", bill_sum_per_day)

# 3. List of tip values greater than 5 using filter()
tips_above_5 = list(filter(lambda x: x > 5, df['tip']))
print("\nTips greater than 5:", tips_above_5)

# 4. Sum of all tips using reduce()
tip_sum = reduce(lambda x, y: x + y, df['tip'])
print("\nSum of all tips:", tip_sum)
```

```python
import pandas as pd

data = {
    'product_name': ['TV', 'Laptop', 'Shirt', 'Blender', 'Jeans', 'Mixer', 'Sofa', 'Tâ€‘shirt'],
    'category': ['electronics', 'electronics', 'clothing', 'home and kitchen',
                 'clothing', 'home and kitchen', 'home and kitchen', 'clothing'],
    'price': [400, 1200, 50, 200, 80, 150, 800, 30],
    'quantity_sold': [5, 3, 20, 7, 15, 4, 2, 25]
}

df = pd.DataFrame(data)
pivot_avg = pd.pivot_table(
    df,
    index='category',
    values='price',
    aggfunc='mean'
)
print(pivot_avg)
df['revenue'] = df['price'] * df['quantity_sold']
pivot_revenue_price = pd.pivot_table(
    df,
    index='category',
    values=['revenue', 'price'],
    aggfunc={'revenue': 'sum', 'price': 'mean'}
)
print(pivot_revenue_price)
```

```python
import pandas as pd
import numpy as np
sr1=pd.Series([1,2,3,4,5])
sr2=pd.Series([2,4,6,8,10])
print("original series")
print("sr1:")
print(sr1)
print("sr2:")
print(sr2)
print("\n items of a given series:")
sr11=pd.Series(np.union1d(sr1,sr2))
sr22=pd.Series(np.intersect1d(sr1,sr2))
print(sr11)
print(sr22)
result=sr11[~sr11.isin(sr22)]
print(result)
```
